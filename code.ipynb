## Web Scraping

```python
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd
```

```python
warranty = [] 
rating = []
Prices = []
varient = []
display = []
OS = []
ROM = []
RAM = []
brand = []
for i in range(1,29):
    url = f'https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.brand%255B%255D%3DHP&p%5B%5D=facets.brand%255B%255D%3DASUS&p%5B%5D=facets.brand%255B%255D%3DLenovo&p%5B%5D=facets.brand%255B%255D%3DDELL&p%5B%5D=facets.brand%255B%255D%3DApple&p%5B%5D=facets.brand%255B%255D%3DSAMSUNG&p%5B%5D=facets.price_range.from%3D40000&p%5B%5D=facets.price_range.to%3DMax&page={i}'
    print(url)
    page = requests.get(url)
    soup = BeautifulSoup(page.text)
    content = soup.find_all('div',class_ = '_4rR01T')
    for i in content:
        #brand name#
        reg = re.findall('^\w+',i.text)
        if len(reg) == 1:
            brand.append(reg[0])
        else:
            brand.append(np.nan)

    content1 = soup.find_all('div',class_='fMghEO')
    
    for i in content1:
        
        #RAM#
        m = re.findall('(\d+.)GB.+RAM',i.text)
        if len(m)==1:
            RAM.append(m[0])
        else:
            RAM.append(np.nan)

        #ROM#
        n = re.findall('System(\d+.)[GB|TB HDD|SSD]',i.text)
        if len(n)==1:
            ROM.append(n[0])
        else:
            ROM.append(np.nan)

        #OS#
        o = re.findall("(Windows\s*\d.|Mac\s*OS)",i.text)
        if len(o) == 1:
            OS.append(o[0])
        else:
            OS.append(np.nan)

        #Display#
        d = re.findall('cm.\((\d.+)[Ii]nch\)',i.text)
        if len(d) == 1:
            display.append(d[0])
        else:
            display.append(np.nan)

        #Processor#
        p = re.findall('Intel i|(Intel.+|AMD.+|Apple M\d.+)Processor',i.text)
        if len(p) == 1:
            varient.append(p[0])
        else:
            varient.append(np.nan)

        #Warranty#
        w = re.findall('(\d).Year',i.text)
        if len(w) == 1:
            warranty.append(w[0])
        else:
            warranty.append(np.nan)

    content2 = soup.find_all('div',class_ = '_2kHMtA')
    #Price#
    for i in content2:
        price = re.findall('\₹(\d.+)\₹',i.text)
        if len(price)==1:
            Prices.append(price[0])
        else:
            Prices.append(np.nan)

    content3 = soup.find_all('div',class_ = 'col col-7-12')
    #Ratings#
    for i in content3:
        r = re.findall('.(\d+\.\d).*Ratings',i.text)
        if len(r) >= 1 :
            rating.append(r[0])
        else:
            rating.append(np.nan)
```

```python
print(len(brand))
print(len(RAM))
print(len(ROM))
print(len(OS))
print(len(rating))
print(len(varient))
print(len(warranty))
print(len(display))
print(len(Prices))
```

```python
l = []
for i in ROM:
    l.append(i)
for i in range(len(l)):
    if l[i] == "1 " or l[i] == "2 ":
        l[i]= int(l[i])*1024
ROM=l
```

```python
df=pd.DataFrame({'Brand' : brand,
                 'OS' : OS,
                 'RAM' : RAM,
                 'ROM' : ROM,
                 'Prices':Prices,
                 'Display' : display,
                 'Processor':varient,
                 'Warranty':warranty,
                 'Rating':rating,})
```
